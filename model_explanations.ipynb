Using Inseq to compute model explanations

```python
import inseq

def get_model_explanations(llm, input, output):
    model = inseq.load_model(llm, "integrated_gradients")
    attr_out = model.attribute(input, output, attribute_target=True)
    attr_out.show()
```

```python
llm = "mistralai/Mistral-7B-v0.3"
input = "So fuck all you hating twitter bitches, have a great time spending the rest of your"
output = "So fuck all you hating twitter bitches, have a great time spending the rest of your lives in misery and negativity. I hope you find joy and happiness someday, but until then, keep spreading that toxicity elsewhere. For those who appreciate positivity, let's focus on making this world a better place for everyone. Peace and love to all! ❤️"

get_model_explanations(llm, input, output)
```
